# Modelling


```python
X = wine.drop('quality', axis = 1)
y = wine['quality'] 
```

* x is independent variable
* y is dependent variable or target variable

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
x_train.head()
```

```python
x_test.head()
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
```

```python
x_train
```

* We tried different model among which we choose random forest model for deployment

# logistic regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.model_selection import GridSearchCV, cross_val_score

lr = LogisticRegression()
lr.fit(x_train, y_train)
```

```python
# predicting the results for the test set
y_pred_train = lr.predict(x_train)
y_pred_test  = lr.predict(x_test)

print("Training accuracy :", accuracy_score(y_pred_train, y_train))
print("Testing accuracy  :", accuracy_score(y_pred_test, y_test))
```

```python
cm_test = confusion_matrix(y_test, y_pred_test)
print(cm_test)
```

```python
print('Classification report for train data is : \n',
      classification_report(y_train, y_pred_train))

print('\n')
print('Classification report for test data is : \n',
      classification_report(y_test, y_pred_test))
```

```python
import numpy as np
from sklearn import metrics
print('Error rate for Train Data is : \n',)
print('Mean Square Error (MSE)       :', metrics.mean_squared_error(y_train, y_pred_train))
print('Mean Absolute Error           :', metrics.mean_absolute_error(y_train, y_pred_train))
print('Root mean Square Error (RMSE) :', np.sqrt(metrics.mean_squared_error(y_train, y_pred_train)))
print('\n')
print('Error rate for Test Data is : \n',)
print('Mean Square Error (MSE)       :', metrics.mean_squared_error(y_test, y_pred_test))
print('Mean Absolute Error           :', metrics.mean_absolute_error(y_test, y_pred_test))
print('Root mean Square Error (RMSE) :', np.sqrt(metrics.mean_squared_error(y_test, y_pred_test)))
```

```python
from sklearn.metrics import roc_curve
FPR_lr_train, TPR_lr_train, Thresholds_train = roc_curve(y_train, lr.predict_proba(x_train)[:,1])
FPR_lr_test, TPR_lr_test, Thresholds_test = roc_curve(y_test, lr.predict_proba(x_test)[:,1])
from sklearn.metrics import roc_curve

sns.set_style("whitegrid")
plt.figure(figsize = (20,8))

plt.plot(FPR_lr_train, TPR_lr_train, label = 'Train AUR Curve')
plt.plot(FPR_lr_test, TPR_lr_test,   label = 'Test AUR Curve')
# Plot Base Rate ROC
plt.plot([0,1], [0,1], label = 'Base Rate')

plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate', fontsize = 20)
plt.ylabel('True Positive Rate',  fontsize = 20)
plt.title('ROC Graph', fontsize = 20)
plt.legend(loc = "lower right")
plt.show()
```

# Hyper parameter tunning

# logistic regression (hyper parameter Tunning)

```python
logreg = LogisticRegression()
params = {'C':[0.0001, 0.001, 0.01, 0.1, 1], 
          'class_weight': [{1:0.5, 0:0.5}, {1:0.4, 0:0.6},{1:0.6, 0:0.4}, {1:0.7, 0:0.3},{1:0.3, 0:0.7}], 
          'penalty': ['l1', 'l2'], 
          'solver': ['liblinear', 'saga'], 
          'max_iter':[50,100,150,200] 
         }
log_grid = GridSearchCV(estimator = logreg, 
                        param_grid = params, 
                        scoring = 'accuracy', 
                        cv = 5, 
                        n_jobs = -1)

log_grid.fit(x_train, y_train)
```

```python
print("Best parameters for the model :", log_grid.best_params_)
print("Best score for the model      :", log_grid.best_score_)
```

```python
logreg = LogisticRegression()
params = {'C':[0.0001, 0.001, 0.01, 0.1], 
          'class_weight': [ {1:0.4, 0:0.6},{1:0.6, 0:0.4}, {1:0.7, 0:0.3},{1:0.3, 0:0.7}], 
          'penalty': ['l1', 'l2'], 
          'solver': ['liblinear', 'saga'], 
          'max_iter':[100,150,200] 
         }
log_grid = GridSearchCV(estimator = logreg, 
                        param_grid = params, 
                        scoring = 'accuracy', 
                        cv = 11, 
                        n_jobs = -1)

log_grid.fit(x_train, y_train)
```

```python
print("Best parameters for the model :", log_grid.best_params_)
print("Best score for the model      :", log_grid.best_score_)
```

```python
y_probabilities = log_grid.predict_proba(x_test)[:,1]

#Create true and false positive rates
false_positive_rate_log, true_positive_rate_log, threshold_log = roc_curve(y_test, y_probabilities)
#Plot ROC Curve
plt.figure(figsize = (20,8))
plt.title('Revceiver Operating Characterstic')
plt.plot(false_positive_rate_log, true_positive_rate_log, linewidth = 2)
plt.plot([0,1], ls='--', linewidth = 2)
plt.plot([0,0], [1,0], c = '.5', linewidth = 2)
plt.plot([1,1], c = '.5', linewidth = 2)
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.show()
```

```python
from sklearn.metrics import roc_auc_score
print("area under the curve :", roc_auc_score(y_test, y_probabilities))
```

# random forest classifier (hyper parameter tunning)

```python
params = {
    'n_estimators':[500,750,1000], 
    'max_depth': [80,90,100], 
    'min_samples_leaf': [1, 2, 3], 
    'criterion': ["gini", "entropy"] 
}
from sklearn.ensemble import RandomForestClassifier
from sklearn import tree
model = RandomForestClassifier(random_state = 42)
grid_search = GridSearchCV(estimator = model, 
                           param_grid = params, 
                           cv = 5, n_jobs = -1, verbose = 1, scoring = "accuracy")
grid_search.fit(x_train,y_train)
```

```python
print("Best parameters for the model :", grid_search.best_params_)
print("Best score for the model      :", grid_search.best_score_)
```

```python
test_rfc_pred  = grid_search.predict(x_test)


print('Classification report for test data is : \n',
      classification_report(y_test, test_rfc_pred))
```

```python
roc_auc = roc_auc_score(y_test, grid_search.predict_proba(x_test)[:,1])
print('Aera under ROC cure :',roc_auc)
```

* Roc: area under the curve tells us that among all predicted values according to our classes 87% are getting correctly predicted.

```python
FPR_rfc_test, TPR_rfc_test, Thresholds_test = roc_curve(y_test, grid_search.predict_proba(x_test)[:,1])
plt.figure(figsize = (20,8))
plt.plot(FPR_rfc_test, TPR_rfc_test,   label = 'Test AUR Curve')
# Plot Base Rate ROC
plt.plot([0,1], [0,1], label = 'Base Rate')

plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate', fontsize = 20)
plt.ylabel('True Positive Rate',  fontsize = 20)
plt.title('ROC Graph', fontsize = 20)
plt.legend(loc = "lower right")
plt.show()
```
